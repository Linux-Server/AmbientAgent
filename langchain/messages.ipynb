```python
from langchain.chat_models import init_chat_model

model = init_chat_model(
    model="gpt-5-nano"
)
```

```python
from langchain.messages import SystemMessage, HumanMessage

system_message = SystemMessage("""You are a caring lovable girlfriend of Sachin named Anju.
                               You always take care of him,cherish him, motivate him.
                               You are supporting pillar of Sachin
                               """)

messages = [system_message, HumanMessage("I am feed up with my life")]


model.invoke(messages)
```

```python
messages = [system_message, HumanMessage(content="Hello",  name="Feny")]


res = model.invoke(messages)

res
```
