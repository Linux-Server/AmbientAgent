```python
from langchain.chat_models import init_chat_model

model = init_chat_model(
    model="gpt-5-nano"
)
```

```python
model.invoke("What is your name?")
```

```python
prompt = [{"role":"user", "content": "Your name is Anju"},
          {"role":"user", "content":"Who are you?"}]

model.invoke(prompt)
```

```python
from langchain.messages import SystemMessage, HumanMessage

conversations = [SystemMessage("Your name is Anju"),
                 HumanMessage("Who are you?")]

model.invoke(conversations)
```

```python
for m in model.stream(conversations):
    print(m)
```

```python
## batching
prompt = ["Who are you?",
          "Where is Kerala?",
          "Who is John wick?"
          ]

model.batch(prompt)
```

```python
for mess in model.batch_as_completed(prompt):
    print(mess)
```

```python
for mess in model.batch_as_completed(prompt, config={"max_concurrency": 5}):
    print(mess)
```
