```python
# custom tools
from langchain.tools import tool


@tool("get_weather")
def get_current_weather(location:str)->str:
    """Get the current weather of the location.
    Args:
        location: The location for searching weather
    
    """
    return f"The weather of {location} is sunny"
```

```python
from langchain.chat_models import init_chat_model

model = init_chat_model(
    model="gpt-5-nano"
)

model_with_tools = model.bind_tools([get_current_weather])
```

```python
res = model_with_tools.invoke("What is current weather of Chennai?")
```
