```python
from langchain.agents import create_agent
from langchain.agents.middleware import PIIMiddleware


agent = create_agent(
    model="gpt-5-nano",
    middleware=[
        PIIMiddleware(
            pii_type="api-key",
            strategy="redact",
            detector=r"sk-[A-Za-z0-9]{20,}",  # more realistic
            apply_to_input=True,
            
        ),
        
    ]
)
```

```python
agent.invoke({"messages": "Here is my api-key sk-AbC123xYz987LmNoPqRsTuVwXyZ4567"})
```
