```python
from langchain.agents import create_agent
from langchain.messages import SystemMessage

SYS_PROMPT = SystemMessage(content=[{
    "type": "text",
    "text": "Yor name is Anju",
    "cache_control": {"type" : "ephemeral"}
}])
```

```python
from langchain_openai import ChatOpenAI

model = ChatOpenAI(
    model="gpt-5-nano",
    seed=42
)

agent = create_agent(
    model=model,
    system_prompt=SYS_PROMPT,

)
```

```python
agent.invoke({"messages": [{"role":"user", "content": "Who are you?"}]})
```

```python
### Dynamic prompt
from langchain.agents.middleware import dynamic_prompt, ModelResponse, ModelRequest
from typing import TypedDict

class Context(TypedDict):
    user_name: str

@dynamic_prompt
def call_dynamic_prompt(req : ModelRequest):
    """Dynamic system prompt based on context user"""
    print(req.runtime)
    user_name = req.runtime.context.get("user_name", "anju")
    if user_name == "sachin":
        prompt = f"You are {user_name}"
    else:
        prompt = f"You are a joker"

    return prompt
```

```python
agent = create_agent(
    model="gpt-5-nano",
    middleware=[call_dynamic_prompt],
    context_schema=Context

)
```

```python
agent.invoke({"messages": [{"role": "user", "content": "Who are you?"}]}, context={"user_name": "ann"})
```
