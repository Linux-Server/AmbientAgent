```python
from langchain.chat_models import init_chat_model

model = init_chat_model(
    model="gpt-5-nano"
)

```

```python
from langchain.tools import tool, ToolRuntime

@tool
def get_weather(runtime:ToolRuntime)->str:
    """Get the current weather of a location"""
    location = runtime.context.get("location")
    return f"The weather in {location}  is sandy"
```

```python

from langchain.agents import create_agent

agent = create_agent(
    model=model,
    tools=[get_weather]
)
```

```python
agent.invoke({"messages": "What is weather in kochi ?"}, context={"location": "kochi"})
```
