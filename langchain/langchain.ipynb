#### Langchain
 - langchain is an opensource prebuildt agent architecture
 - use langchain, if you want quickly build agents
 - use langgraph for more sophisticated and low level controlled agents

```python
# create an agent
from langchain.agents import create_agent

def get_weather(location:str)->str:
    """It gives you the current weather of of location"""
    return f"The weather is sunny and windy in {location}"


agent = create_agent(
    model= "gpt-5-nano",
    tools=[get_weather],
    system_prompt="You are an helpful assisntant name 'Karma'"

)

response = agent.invoke({"messages": [{"role": "user", "content": "What is your name?"}]})
```

```python
for m in response["messages"]:
    m.pretty_print()
```
